## What is Basilisk?

Basilisk (BSK) is an **open-source, modular astrodynamics simulation framework** written primarily in C/C++ with a Python scripting interface via SWIG. Think of it as a set of spacecraft *Lego blocks* — you assemble them into the spacecraft behavior you want.

### Key Characteristics
- **Modular**: Every behavior is a standalone module with message-passing I/O
- **Fast**: C/C++ core enables faster-than-realtime simulation
- **Scriptable**: Full Python interface for setup, control, and analysis
- **Validated**: Each module has unit tests; integrated tests validate combined behavior
- **Visualizable**: Companion tool *Vizard* (Unity-based) renders spacecraft and orbits

### Use Cases
- Orbit and attitude dynamics research
- Flight software (FSW) algorithm development and testing
- Monte Carlo analysis
- Hardware-in-the-loop (HIL) simulation
- Mission design and validation

## Architecture Overview

```
┌─────────────────────────────────────────────────────────────┐
│                  Python Scripting Layer                     │
│          (Setup, Configuration, Data Analysis)              │
└───────────────────────┬─────────────────────────────────────┘
                        │ SWIG Interface
┌───────────────────────▼─────────────────────────────────────┐
│                  SimulationBaseClass                        │
│  ┌──────────────┐  ┌──────────────┐  ┌──────────────────┐  │
│  │   Process 1  │  │   Process 2  │  │    Process N     │  │
│  │  ┌────────┐  │  │  ┌────────┐  │  │  ┌────────────┐  │  │
│  │  │ Task A │  │  │  │ Task C │  │  │  │   Task X   │  │  │
│  │  │ mod1   │  │  │  │ mod3   │  │  │  │   modN     │  │  │
│  │  │ mod2   │  │  │  │ mod4   │  │  │  │   ...      │  │  │
│  │  └────────┘  │  │  └────────┘  │  │  └────────────┘  │  │
│  │  ┌────────┐  │  │              │  │                   │  │
│  │  │ Task B │  │  │              │  │                   │  │
│  │  └────────┘  │  │              │  │                   │  │
│  └──────────────┘  └──────────────┘  └──────────────────┘  │
│                                                             │
│              Message Bus (Topic-based I/O)                  │
└─────────────────────────────────────────────────────────────┘
```

**Processes** → containers that group related **Tasks**  
**Tasks** → containers that group related **Modules** and define an execution rate  
**Modules** → atomic C/C++ units that read input messages and write output messages  
**Messages** → the typed data structures modules use to communicate

## Basilisk Package Structure

After installation, Basilisk exposes the following top-level sub-packages:

| Sub-package | Contents |
|---|---|
| `Basilisk.simulation` | Dynamics modules (spacecraft, gravity, sensors, actuators, …) |
| `Basilisk.fswAlgorithms` | Flight software algorithm modules (attitude guidance, control, …) |
| `Basilisk.utilities` | Python helpers — `SimulationBaseClass`, `macros`, `orbitalMotion`, `RigidBodyKinematics`, … |
| `Basilisk.architecture` | Core message types (`messaging`), `bskLogging` |
| `Basilisk.topLevelModules` | Higher-level containers like `Spacecraft` |
| `Basilisk.externalTools` | Third-party integrations (e.g. `spiceInterface`) |
